## Tarea 3
Muestra la masa total de carga útil transportada por los propulsores lanzados por la NASA (CRS)

```python
%sql SELECT SUM(PAYLOAD_MASS__KG_) AS TOTAL_PAYLOAD_MASS__KG FROM SPACEXTBL WHERE  Customer LIKE  '%NASA (CRS)%';
```

## Tarea 4
Muestra la masa de carga útil promedio transportada por la versión de refuerzo F9 v1.1

```python
%sql SELECT AVG(PAYLOAD_MASS__KG_) AS MASS_MEAN FROM SPACEXTBL WHERE Booster_Version = 'F9 v1.1';
```

## Tarea 5
Indique la fecha en la que se logró el primer aterrizaje exitoso en la plataforma de tierra.

```python
%sql SELECT MIN(Date) AS primer_aterrizaje_exitoso FROM SPACEXTBL WHERE Landing_Outcome LIKE '%Controlled%';
```

## Tarea 6
Enumere los nombres de los propulsores que tienen éxito en naves no tripuladas y tienen una masa de carga superior a 4000 pero inferior a 6000.

```python
%sql SELECT Booster_Version FROM SPACEXTBL WHERE Landing_Outcome LIKE '%drone ship%' OR '%Controlled%' AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000;
```

## Tarea 7
Enumere el número total de resultados de misiones exitosas y fallidas.

## Tarea 8
Enumere los nombres de las booster_versions que han transportado la masa máxima de carga útil. Utilice una subconsulta

## Tarea 9
Enumere los registros que mostrarán los nombres de los meses, los resultados de aterrizaje fallidos en el barco no tripulado, las versiones de refuerzo y el sitio de lanzamiento para los meses del año 2015.
Nota: SQLLite no admite nombres de meses. Por lo tanto, debe usar substr(Date, 6,2) como mes para obtener los meses y substr(Date,0,5)='2015' para el año.

```python
%sql SELECT\
 CASE substr(Date,6,2)\
    WHEN '01' THEN 'Enero'\
    WHEN '02' THEN 'Febrero'\
    WHEN '03' THEN 'Marzo'\
    WHEN '04' THEN 'Abril'\
    WHEN '05' THEN 'Mayo'\
    WHEN '06' THEN 'Junio'\
    WHEN '07' THEN 'Julio'\
    WHEN '08' THEN 'Agosto'\
    WHEN '09' THEN 'Septiembre'\
    WHEN '10' THEN 'Octubre'\
    WHEN '11' THEN 'Noviembre'\
    WHEN '12' THEN 'Diciembre'\
  END as month, DATE,BOOSTER_VERSION, LAUNCH_SITE, [Landing_Outcome] \
FROM SPACEXTBL \
where [Landing_Outcome] = 'Failure (drone ship)' and substr(Date,0,5)='2015';
```

## Tarea 10
Clasifique el recuento de resultados de aterrizaje (como Fracaso (nave no tripulada) o Éxito (plataforma terrestre)) entre la fecha 2010-06-04 y 2017-03-20, en orden descendente.

```python
%sql\
SELECT Landing_Outcome, COUNT(*) AS Aterrizaje\
FROM SPACEXTBL\
WHERE  (Landing_Outcome LIKE '%Success (ground pad)%')  OR (Landing_Outcome LIKE '%Failure (drone ship)%')\
AND Date >='2010-06-04' AND Date <='2017-03-20'\
GROUP BY Landing_Outcome\
ORDER BY  Aterrizaje DESC;
```

```python
%sql SELECT Landing_Outcome FROM(Landing_Outcome LIKE '%Success (ground pad)%'),\
COUNT(*) AS EXITO, FROM (Landing_Outcome LIKE '%Failure (drone ship)%'), COUNT AS FAILURE\
FROM SPACEXTBL\
WHERE Date >='2010-06-04' AND Date <='2017-03-20'\
GROUP BY Landing_Outcome\
ORDER BY  Aterrizaje DESC;
```
